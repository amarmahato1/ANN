# Importing Libraries 

```python
# Import necessary libraries
# Linear algebra and data processing
import numpy as np
import pandas as pd

# Data visualization
import matplotlib.pyplot as plt

# Mathematical functions
import math

# Deep learning libraries
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

# Data preprocessing
from sklearn.preprocessing import MinMaxScaler

# Metrics for evaluation
from sklearn.metrics import mean_squared_error

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

# Data Preparation

```python
raw_data.info()
```

```python
# Check for missing values in the 'raw_data' DataFrame
raw_data.isna().sum()
```

```python
# Remove rows with missing values from the 'raw_data' DataFrame
raw_data = raw_data.dropna()
```

```python
raw_data.isna().sum()
```

```python
# Convert the 'Timestamp' column to a datetime format with unit 's' (seconds) and extract the date
raw_data['date'] = pd.to_datetime(raw_data['Timestamp'], unit='s').dt.date
raw_data
```

```python
# Display information about the 'raw_data' DataFrame
raw_data.info()
```

```python
# Group the 'raw_data' DataFrame by the 'date' column and calculate the mean for each group
grouped_data = raw_data.groupby('date', as_index=False).mean()
grouped_data
```

# Train and Test Split

```python
# Calculate the size of the training set as 95% of the total data
train_size = int(len(grouped_data) * 0.95)

# Calculate the size of the test set as the remaining data
test_size = len(grouped_data) - train_size

# Create a training DataFrame by selecting the first 'train_size' rows from 'grouped_data'
df_train = grouped_data.iloc[0:train_size, :]

# Create a test DataFrame by selecting the rows from 'train_size' to the end of 'grouped_data'
df_test = grouped_data.iloc[train_size:len(grouped_data), :]

# Print the sizes of the training and test sets
print("Train size: {}, Test size: {}".format(len(df_train), len(df_test)))
```

```python
# Extract the "Close" column from the training set DataFrame 'df_train' and convert it to a NumPy array
dataset = df_train.loc[:, ["Close"]].values
dataset.shape
```

# Scaling Data

```python
# Import the MinMaxScaler class from scikit-learn
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object with a feature range of (0, 1)
scaler = MinMaxScaler(feature_range=(0, 1))

# Use the scaler to transform (scale) the 'dataset' values and store the result in 'data_scaled'
data_scaled = scaler.fit_transform(dataset)
data_scaled
```

```python
# Create a line plot of the scaled data
plt.plot(data_scaled)

# Set the x-axis label
plt.xlabel("Time")

# Set the y-axis label
plt.ylabel("Values of Close")

# Set the title of the plot
plt.title("Bitcoin Close Price (Scaled)")

# Display the plot
plt.show()
```

```python
# Initialize empty lists to store the training data and labels
X_train = []  # Training data
y_train = []  # Training labels

# Define the time step, which represents the number of time steps to consider for each sample
time_step = 10

# Iterate over the scaled data to create training samples
for i in range(time_step, data_scaled.shape[0]):
    # Create input sequences with 'time_step' time steps
    X_train.append(data_scaled[i - time_step:i, 0])
    # Create corresponding output labels
    y_train.append(data_scaled[i, 0])

# Convert the lists to NumPy arrays
X_train, y_train = np.array(X_train), np.array(y_train)

# Print the shapes of the training data and labels
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
```

```python
# Reshape the 'X_train' array to include an additional dimension (1) to indicate the number of features
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```
