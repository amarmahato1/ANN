# Data pre-processing

```python
#There are some points with outliers, let's change to NaN and then call a method to fill the empty values
data['Weighted_Price'].replace(0, np.nan, inplace=True)
data['Weighted_Price'].fillna(method='ffill', inplace=True)
```

```python
# Let's use the weighted price as input to our recurring network
# Normalizes in the range between [0 and 1]
values = data['Weighted_Price'].values.reshape(-1,1)
values = values.astype('float32')
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
```

Split data in train and test

```python
# Calculate the training size as 70% of the total length of the 'scaled' data
train_size = int(len(scaled) * 0.7)

# Calculate the test size as the remaining data after the training set
test_size = len(scaled) - train_size

# Split the 'scaled' data into training and test sets
# 'train' will contain the first 'train_size' rows, and 'test' will contain the remaining rows
train, test = scaled[0:train_size, :], scaled[train_size:len(scaled), :]

# Print the lengths of the training and test sets to verify the split
print(len(train), len(test))
```

```python
# Define a function called 'create_dataset' that takes 'dataset' and 'look_back' as parameters
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []  # Initialize empty lists to store input (dataX) and output (dataY)

    # Iterate through the 'dataset' to create input-output pairs
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]  # 'a' contains a sequence of 'look_back' values from 'dataset'
        dataX.append(a)  # Append 'a' to 'dataX' as input
        dataY.append(dataset[i + look_back, 0])  # Append the next value as the corresponding output

    # Print the number of data pairs created
    print(len(dataY))

    # Return 'dataX' and 'dataY' as NumPy arrays representing the input and output sequences
    return np.array(dataX), np.array(dataY)
```

```python
# Enter with 1 single value window (Just one day)
look_back = 1  # Set the look_back parameter to 1, meaning we use one past time step as input

# Create training input-output pairs using the 'create_dataset' function
trainX, trainY = create_dataset(train, look_back)

# Create testing input-output pairs using the 'create_dataset' function
testX, testY = create_dataset(test, look_back)
```

```python
# reshape to neural network input format (instances, 1, 1)
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
```

# LSTM Model

```python
# Create a Sequential model, which is a linear stack of layers
model = Sequential()

# Add a Bidirectional LSTM layer with 32 units
model.add(Bidirectional(LSTM(32)))

# Add a fully connected layer with 128 units and ReLU activation
model.add(Dense(units=128, activation='relu'))

# Add a dropout layer with a dropout rate of 0.5 to prevent overfitting
model.add(Dropout(0.5))

# Add another fully connected layer with 64 units and ReLU activation
model.add(Dense(units=64, activation='relu'))

# Add another dropout layer with a dropout rate of 0.5
model.add(Dropout(0.5))

# Add a final output layer with 1 unit (for regression tasks)
model.add(Dense(units = 1))

# Build the model specifying the input shape based on the dimensions of 'trainX'
model.build(input_shape= (trainX.shape[0], trainX.shape[1], trainX.shape[2]))

model.summary()
```

Function for training

```python
# Define a function 'train_model' that takes a neural network model, loss, optimizer, training data, and labels as parameters
def train_model(model, loss, optimizer, trainX, trainY):
    # Compile the model with the specified loss function and optimizer
    model.compile(loss=loss, optimizer=optimizer)
    
    # Train the model using the training data and labels
    history = model.fit(trainX, trainY, epochs=30, batch_size=200, validation_data=(testX, testY))
    
    # Plot the training and testing loss curves to visualize the model's learning progress
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.legend()
    plt.show()
```

Function to evaluate model

```python
# Define a function 'evaluate_model' that takes a trained model, test data, and scaler as parameters
def evaluate_model(model, testX, testY):
    # Use the trained model to make predictions on the test data
    yhat = model.predict(testX)
    
    # Plot the actual and predicted values on the same scale used during training
    plt.title('Curve of actual value and predicted value on the scale used in training')
    plt.plot(yhat, label='predict')
    plt.plot(testY, label='true')
    plt.legend()
    plt.show()
    
    # Inverse transform the scaled predictions and true values to the original scale
    yhat_inverse = scaler.inverse_transform(yhat.reshape(-1, 1))
    testY_inverse = scaler.inverse_transform(testY.reshape(-1, 1))
    
    # Calculate the root mean squared error (RMSE) between the actual and predicted values
    rmse = sqrt(mean_squared_error(testY_inverse, yhat_inverse))
    print('Test RMSE: %.3f' % rmse)

    # Plot the actual and predicted values in US dollars
    plt.title('Actual value curve and predicted value in US dollars')
    plt.plot(yhat_inverse, label='predict')
    plt.plot(testY_inverse, label='actual', alpha=0.5)
    plt.legend()
    plt.show()
```

# Training and evaluating model

```python
# Train the 'model' using the 'train_model' function with specified parameters
train_model(model, 'mse', 'adam', trainX, trainY)
```

```python
# test
evaluate_model(model, testX, testY)
```

# Predicting the next day

```python
# Define a function 'create_time_steps' that generates a list of time steps from -length to -1
def create_time_steps(length):
    time_steps = []
    for i in range(-length, 0, 1):
        time_steps.append(i)
    return time_steps

# Define a function 'baseline' that calculates the mean of a history sequence
def baseline(history):
    return np.mean(history)

# Define a function 'show_plot' that displays a plot with history, true future, and model prediction data
def show_plot(plot_data, delta, title):
    labels = ['History', 'True Future', 'Model Prediction']
    marker = ['.-', 'rx', 'go']
    time_steps = create_time_steps(plot_data[0].shape[0])
    
    if delta:
        future = delta
    else:
        future = 0

    plt.title(title)
    for i, x in enumerate(plot_data):
        if i:
            plt.plot(future, plot_data[i], marker[i], markersize=10, label=labels[i])
        else:
            plt.plot(time_steps, plot_data[i].flatten(), marker[i], label=labels[i])
    
    plt.legend()
    plt.xlim([time_steps[0], (future + 5) * 2])
    plt.xlabel('Time-Step')
    return plt
```

```python
# Define a function 'predict_next_day' that makes predictions for the next day using a trained model and test data
def predict_next_day(model, testX, testY):
    # Inverse transform the scaled predictions and true values to the original scale
    yhat_inverse = scaler.inverse_transform(model.predict(testX).reshape(-1, 1))
    testY_inverse = scaler.inverse_transform(testY.reshape(-1, 1))

    # Display a plot of the true values, predicted value for the next day, and a baseline value
    show_plot([testY_inverse[0:200], yhat_inverse[201], baseline(testY_inverse[201])], 1, 'Next day prediction')
    
    # Print the predicted value for day 201 and the actual value for day 201
    print('Predicted value of day 201: ', yhat_inverse[201])
    print('Actual value of day 201: ', testY_inverse[201])
```

```python
predict_next_day(model, testX, testY)
```
